### Problem 2

```python
! pip install stop-words
```

```python
from stop_words import get_stop_words

all_stop_words = get_stop_words('en')
all_stop_words.remove('not')
all_stop_words.remove('cannot')
all_stop_words.remove('didn\'t')
all_stop_words.append('<br />')

print(all_stop_words)
```

```python
for i in range(len(imdb_mod)):
    text = imdb_mod['review'][i]
    filtered = ' '.join([word for word in text.split() if word not in all_stop_words])
    imdb_mod['filtered_review'][i] = filtered
```

```python
imdb_mod.tail()
```

```python
# convert imdb_mod to a SFrame in order to use the count_words 
df_reviews = tc.SFrame(imdb_mod)
df_reviews


df_reviews['words'] = tc.text_analytics.count_words(df_reviews['filtered_review'])
df_reviews

```

### Problem 2
#### d) [1 mark] Split the data into training/validation/testing sets using 80%/10%/10% respectively.
ans.

```python
# split 80% of the data --> 0.8
train_data, rest_data = df_reviews.random_split(.8, seed=0, exact=True)

# split rest of data in 50% ( 10% validation + 10% test)--> 0.5
validation_data, test_data = rest_data.random_split(.5, seed=0, exact=True)


# check the length of dataset 
print("Length of dataset:", len(df_reviews))

# check the length of training set
print("Length of training set:", len(train_data))

# check the length of validation set
print("Length of validation set:", len(validation_data))

# check the length of testing set
print("Length of testing set:", len(test_data))
```

```python
model = tc.logistic_classifier.create(df_reviews, 
                                      features=['words'], 
                                      target='sentiment',
                                      validation_set=validation_data, 
                                      l1_penalty=0.1, 
                                      l2_penalty=0.01
                                     )
```

```python
model
```

```python
weights = model.coefficients
weights
```

```python
# show negative review words
weights.sort('value')
```

```python
# show positive review words
weights.sort('value', ascending=False)
```

```python
df_reviews['predictions'] = model.predict(df_reviews, output_type='probability')
```

```python
df_reviews.sort('predictions', ascending=False)[0]
```

```python
df_reviews.sort('predictions', ascending=True)[0]
```

```python
# # calculate the accuracy of training data
# accuracy_train = tc.evaluation.accuracy(train_data["words"], model.predict(train_data))

# # calculate the accuracy of validation data
# accuracy_validate = tc.evaluation.accuracy(validation_data["words"], model.predict(validation_data))

result = model.evaluate(test_data)
result

recall_prediction = result.get('recall')
precision_prediction = result.get('precision')
auc = result.get('auc')
accuracy = result.get('accuracy')

TP = result['confusion_matrix'][2]
TP = int(TP['count'])

TN = result['confusion_matrix'][0]
TN = int(TN['count'])

FN = result['confusion_matrix'][3]
FN = int(FN['count'])

FP = result['confusion_matrix'][1]
FP = int(FP ['count'])

sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)

print('Accuracy:', accuracy,
      '\nAUC', auc,
      '\nRecall of model 2:', recall_prediction,
      '\nPrecision of model 2', precision_prediction,
      '\nSensitivity of model 2', sensitivity,
      '\nSpecificity of model 2', specificity)
```
